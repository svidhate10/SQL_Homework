```python
query1 = 'select avg(s.salary)as avs,t.title as title from salaries as s,titles as t where (s.emp_no=t.emp_no) group by t.title;'
q1=connection.execute(query1)
x_axis=[]
y_axis=[]
for row1 in q1:
    print(row1["avs"],row1["title"])
    #x_axis=row1["title"]
    x_axis.append(row1["title"])
    #y_axis=row1["avs"]
    y_axis.append(row1["avs"])
```
